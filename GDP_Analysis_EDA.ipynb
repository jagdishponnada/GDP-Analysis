### Categorise the states into four categories based on GDP per capita (C1, C2, C3, C4 - C1 would have the highest per capita GPD, C4 the lowest). The quantile values are (0.20,0.5, 0.85, 1), i.e. the states lying between the 85th and the 100th percentile are in C1, those between 50th and 85th percentile are in C2 and so on.
#### 1.	Note: Categorisation into four categories will simplify the subsequent analysis, since comparing data of all the states would become quite exhaustive

```python
#calculating the percentile info for the Per capita GDP
import numpy as np
a = np.array(Master_df.per_capita_GDP)
P20=int(np.percentile(a, 20))#returns 20th percentile
P50=int(np.percentile(a, 50)) 
P85=int(np.percentile(a, 85))
P100=int(np.percentile(a, 100)) 
print("The 20th,50th,85th and 100th percentile are " ,P20,P50,P85,P100)
```

```python
#converting the GDP values to int to support boolean operations
Master_df.per_capita_GDP=Master_df.per_capita_GDP.astype(int)
```

```python
#Lets categorize the states as per the Per capits GDP
#Lets add a new column which is C1 - (85 -100),C2-(50-85),C3 -(20-50),C4-(0-20)
def category_func(x):
    if (0<x & x<=P20):
        return "C4"
    elif (P20<x & x<=P50):
        return "C3"
    elif (P50<x & x<=P85):
        return "C2"
    else:
        return "C1"

Master_df['Category']=Master_df.per_capita_GDP.apply(category_func)
```

```python
#creting four different data frames for each category which contains top 7 contributers  and GSDP, as after sorting in 
#descending order the top will be GSDP
C1_df=category_df[['C1']].sort_values(by='C1',ascending=False).head(8)
C2_df=category_df[['C2']].sort_values(by='C2',ascending=False).head(8)
C3_df=category_df[['C3']].sort_values(by='C3',ascending=False).head(8)
C4_df=category_df[['C4']].sort_values(by='C4',ascending=False).head(8)

#calculating the contribution % of each sub sector on the total GSDP

C1_df["%_of_GSDP"]=C1_df.C1/C1_df.loc['GSDP','C1']*100
C2_df["%_of_GSDP"]=C2_df.C2/C2_df.loc['GSDP','C2']*100
C3_df["%_of_GSDP"]=C3_df.C3/C3_df.loc['GSDP','C3']*100
C4_df["%_of_GSDP"]=C4_df.C4/C4_df.loc['GSDP','C4']*100
```

```python
#we can drop CSDP as we not need it anymore
C1_df.drop(['GSDP'],axis=0,inplace=True)
C2_df.drop(['GSDP'],axis=0,inplace=True)
C3_df.drop(['GSDP'],axis=0,inplace=True)
C4_df.drop(['GSDP'],axis=0,inplace=True)
```

```python
#Creating another column cumulative_% which calculates the cumulative sum of the "%_of_GSDP" . if the max of the
#cumulative sum could reaches 80%(almost) then we are good  If no ,then we have to continue by adding more sectors .If overshoots 
#then we will remove some sectors to make it as close as 80%
C1_df['cumulative_%']=C1_df['%_of_GSDP'].cumsum()
C2_df['cumulative_%']=C2_df['%_of_GSDP'].cumsum()
C3_df['cumulative_%']=C3_df['%_of_GSDP'].cumsum()
C4_df['cumulative_%']=C4_df['%_of_GSDP'].cumsum()
print(C4_df['cumulative_%'].max(),C3_df['cumulative_%'].max(),C2_df['cumulative_%'].max(),C1_df['cumulative_%'].max())
```
